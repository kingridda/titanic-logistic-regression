```python
#data preprocessing
#we drop inuseful column
#inplace means change the original var (train) not return new variable

train.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1, inplace=True)
```

```python
train.isna().sum()
```

```python
#replace missing ages by mean of ages
train["Age"].fillna( train["Age"].mean(), inplace=True)
```

```python
#replace missing embarked values by mode
train["Embarked"].fillna( train["Embarked"].mode()[0], inplace=True)
```

```python
train
```

```python
#replace sex column by 0 and 1 instead of strings (use lambda, map)

train['Sex'] = train['Sex'].apply(lambda val: 1 if val == 'male' else 0)
train['Embarked'] = train['Embarked'].map({'S' : 0, 'C': 1, 'Q': 2}) #another way to do that

#train.apply(lambda val: 1 if val['Sex'] == 'male' else 0 )
```

```python
train
```

```python
train.describe()
```

```python
#split train and dev datasets
y = train['Survived']
X = train.drop('Survived', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

#NB: split function make shuffling by default
#NB random state is seed for random. to generate same random values every time
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)

#warning if number of iterations is low
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

lr_acc = accuracy_score(y_test, lr.predict(X_test))

print(f"Training Accuracy of Logistic Regression is {accuracy_score(y_train, lr.predict(X_train))}")
print(f"Test Accuracy of Logistic Regression is {lr_acc}")

print(confusion_matrix(y_test, lr.predict(X_test)))
print(classification_report(y_test, lr.predict(X_test)))
```

```python
lr = LogisticRegression(max_iter=500)
lr.fit(X_train, y_train)
lr_acc = accuracy_score(y_test, lr.predict(X_test))

print(f"Training Accuracy of Logistic Regression is {accuracy_score(y_train, lr.predict(X_train))}")
print(f"Test Accuracy of Logistic Regression is {lr_acc}")

print(confusion_matrix(y_test, lr.predict(X_test)))
print(classification_report(y_test, lr.predict(X_test)))
```
